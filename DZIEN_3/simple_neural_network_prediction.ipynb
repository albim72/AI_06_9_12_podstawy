```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train/255.0, x_test/255.0
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#budowa sieci neuronowej
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10)
])

model.summary()
```

```python
model.compile(optimizer="adam",
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = ['accuracy']
              )
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
#ocena modelu
test_loss, test_acc = model.evaluate(x_test,y_test)
print(f"strata: {test_loss}")
print(f"accuracy: {test_acc}")
```

```python
#predykcja
logits = model.predict(x_test)

print(f"kształt logit: {logits.shape}")
```

```python
#zamian logitów na prawdopodobięnstwa

import numpy as np
y_proba = tf.nn.softmax(logits,axis=1).numpy()

print(f"kształt prawdopodobieństw : {y_proba.shape}")
print(f"przypadek 0 -> {y_proba[0]}")
print(f"przypadek 0 -> {np.argmax(y_proba[0])}")
print(f"test -> {y_test[0]}")

print(f"przypadek 121 -> {y_proba[121]}")
print(f"przypadek 121 -> {np.argmax(y_proba[121])}")
print(f"test -> {y_test[121]}")

print(f"przypadek 7891 -> {y_proba[7891]}")
print(f"przypadek 7891 -> {np.argmax(y_proba[7891])}")
print(f"test -> {y_test[7891]}")
```

```python
# analiza błędnych predykcji modelu
# Szukamy wszystkich przypadków, w których predykcja modelu jest niezgodna z prawdziwą etykietą.

import numpy as np
import pandas as pd

y_pred = np.argmax(y_proba, axis=1)

# True tam, gdzie model się pomylił
wrong_mask = y_pred != y_test

# indeksy błędnie sklasyfikowanych obrazów
wrong_indices = np.where(wrong_mask)[0]

print(f"Liczba błędnych predykcji: {len(wrong_indices)}")
print(f"Liczba wszystkich przypadków testowych: {len(y_test)}")
print(f"Odsetek błędów: {len(wrong_indices) / len(y_test) * 100:.2f}%")

wrong_cases = pd.DataFrame({
    "indeks": wrong_indices,
    "etykieta_prawdziwa": y_test[wrong_indices],
    "predykcja_modelu": y_pred[wrong_indices],
    "pewność_predykcji": y_proba[wrong_indices, y_pred[wrong_indices]],
    "prawdopodobieństwo_etykiety_prawdziwej": y_proba[wrong_indices, y_test[wrong_indices]]
})

wrong_cases["różnica_pewności"] = (
    wrong_cases["pewność_predykcji"] -
    wrong_cases["prawdopodobieństwo_etykiety_prawdziwej"]
)

wrong_cases = wrong_cases.sort_values(
    by="pewność_predykcji",
    ascending=False
).reset_index(drop=True)

wrong_cases
```
