```python
class Osoba:
    def __init__(self,imie,nazwisko,wiek,wzrost):
        self.imie = imie
        self.nazwisko = nazwisko
        self.wiek = wiek
        self.wzrost = wzrost

    def info(self):
        return f"osoba: {self.imie} {self.nazwisko}, wiek: {self.wiek}, wzrost: {self.wzrost}"
```

```python
os1 = Osoba("Jan","Kowalski",45,178)
print(os1)
print(os1.info())
```
